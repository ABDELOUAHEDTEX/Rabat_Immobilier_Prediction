```python
import pandas as pd
# Charger les deux fichiers CSV
df_mubawab = pd.read_csv('mubawab_properties.csv')
df_avito = pd.read_csv('avito_properties.csv')

# Afficher les premières lignes pour les visualiser
df_mubawab.head(), df_avito.head()
```

```python
# Sélectionner uniquement les colonnes nécessaires
columns_to_keep = ['price', 'type', 'area', 'rooms', 'bedrooms', 'bathrooms','property_state', 'jardin', 'piscine', 'cuisine_equiped','terrasse', 'garage', 'quartier']

df_mubawab_cleaned = df_mubawab[columns_to_keep]
df_avito_cleaned = df_avito[columns_to_keep]
```

```python
df_mubawab_cleaned['quartier'] = (
    df_mubawab_cleaned['quartier']
    .astype(str)  # Convertit tout en chaîne (gère les NaN)
    .str.replace(r'\s+[àa]\s+Rabat$', '', regex=True)  # Supprime " à Rabat", " a Rabat" en fin de ligne
    .str.strip()  # Enlève les espaces résiduels
)
print(df_mubawab_cleaned['quartier'].unique())
print(df_avito_cleaned['quartier'].unique())
```

```python

# Nettoyer la colonne 'price' pour df_mubawab_cleaned
df_mubawab_cleaned.loc[:, 'price'] = df_mubawab_cleaned['price'].astype(str).str.extract(r'(\d[\d\s]*)')[0]
df_mubawab_cleaned.loc[:, 'price'] = df_mubawab_cleaned['price'].str.replace(' ', '', regex=False)
df_mubawab_cleaned.loc[:, 'price'] = pd.to_numeric(df_mubawab_cleaned['price'], errors='coerce')

# Nettoyer la colonne 'price' pour df_avito_cleaned
df_avito_cleaned.loc[:, 'price'] = df_avito_cleaned['price'].astype(str).str.extract(r'(\d[\d\s]*)')[0]
df_avito_cleaned.loc[:, 'price'] = df_avito_cleaned['price'].str.replace(' ', '', regex=False)
df_avito_cleaned.loc[:, 'price'] = pd.to_numeric(df_avito_cleaned['price'], errors='coerce')
```

```python

# Mubawab uses 'oui' and 'non', Avito uses True/False
df_mubawab_cleaned['jardin'] = df_mubawab_cleaned['jardin'].map({'Oui': 1, 'Non': 0})
df_mubawab_cleaned['piscine'] = df_mubawab_cleaned['piscine'].map({'Oui': 1, 'Non': 0})
df_mubawab_cleaned['cuisine_equiped'] = df_mubawab_cleaned['cuisine_equiped'].map({'Oui': 1, 'Non': 0})
df_mubawab_cleaned['terrasse'] = df_mubawab_cleaned['terrasse'].map({'Oui': 1, 'Non': 0})
df_mubawab_cleaned['garage'] = df_mubawab_cleaned['garage'].map({'Oui': 1, 'Non': 0})

df_avito_cleaned['jardin'] = df_avito_cleaned['jardin'].map({'True': 1, 'False': 0})
df_avito_cleaned['piscine'] = df_avito_cleaned['piscine'].map({'True': 1, 'False': 0})
df_avito_cleaned['cuisine_equiped'] = df_avito_cleaned['cuisine_equiped'].map({'True': 1, 'False': 0})
df_avito_cleaned['terrasse'] = df_avito_cleaned['terrasse'].map({'True': 1, 'False': 0})
df_avito_cleaned['garage'] = df_avito_cleaned['garage'].map({'True': 1, 'False': 0})
```

```python
# Convertir les colonnes en types numériques où applicable
df_mubawab_cleaned['price'] = pd.to_numeric(df_mubawab_cleaned['price'], errors='coerce')
df_mubawab_cleaned['area'] = pd.to_numeric(df_mubawab_cleaned['area'], errors='coerce')
df_mubawab_cleaned['rooms'] = pd.to_numeric(df_mubawab_cleaned['rooms'], errors='coerce')
df_mubawab_cleaned['bedrooms'] = pd.to_numeric(df_mubawab_cleaned['bedrooms'], errors='coerce')
df_mubawab_cleaned['bathrooms'] = pd.to_numeric(df_mubawab_cleaned['bathrooms'], errors='coerce')

df_avito_cleaned['price'] = pd.to_numeric(df_avito_cleaned['price'], errors='coerce')
df_avito_cleaned['area'] = pd.to_numeric(df_avito_cleaned['area'], errors='coerce')
df_avito_cleaned['rooms'] = pd.to_numeric(df_avito_cleaned['rooms'], errors='coerce')
df_avito_cleaned['bedrooms'] = pd.to_numeric(df_avito_cleaned['bedrooms'], errors='coerce')
df_avito_cleaned['bathrooms'] = pd.to_numeric(df_avito_cleaned['bathrooms'], errors='coerce')
```

```python
# Compléter les valeurs manquantes pour 'area'
df_mubawab_cleaned['area'] = df_mubawab_cleaned['area'].fillna(
    df_mubawab_cleaned.groupby(['type', 'quartier', 'property_state', 'bedrooms', 'bathrooms'])['area']
    .transform(lambda x: x.mean()))

df_avito_cleaned['area'] = df_avito_cleaned['area'].fillna(
    df_avito_cleaned.groupby(['type', 'quartier', 'property_state', 'bedrooms', 'bathrooms'])['area']
    .transform(lambda x: x.mean()))
```

```python
# Compléter les valeurs manquantes pour 'price'
df_mubawab_cleaned['price'] = df_mubawab_cleaned['price'].fillna(
    df_mubawab_cleaned.groupby(['type', 'quartier', 'property_state', 'bedrooms', 'bathrooms', 'rooms', 'jardin', 'piscine', 'cuisine_equiped'])['price']
    .transform(lambda x: x.mean()))

df_avito_cleaned['price'] = df_avito_cleaned['price'].fillna(
    df_avito_cleaned.groupby(['type', 'quartier', 'property_state', 'bedrooms', 'bathrooms', 'rooms', 'jardin', 'piscine', 'cuisine_equiped'])['price']
    .transform(lambda x: x.mean()))
```

```python
#  Supprimer les lignes avec des valeurs manquantes restantes
df_mubawab_cleaned.dropna(subset=['price', 'area'], inplace=True)
df_avito_cleaned.dropna(subset=['price', 'area'], inplace=True)
```

```python

# Calculer le prix par m²
df_mubawab_cleaned['price_de_m2'] = df_mubawab_cleaned['price'] / df_mubawab_cleaned['area']
df_avito_cleaned['price_de_m2'] = df_avito_cleaned['price'] / df_avito_cleaned['area']

# Réorganiser les colonnes
df_mubawab_cleaned = df_mubawab_cleaned[['type', 'area', 'property_state', 'rooms', 'bedrooms', 'bathrooms', 'jardin', 'piscine', 'cuisine_equiped', 'terrasse', 'garage', 'quartier', 'price_de_m2', 'price']]
df_avito_cleaned = df_avito_cleaned[['type', 'area', 'property_state', 'rooms', 'bedrooms', 'bathrooms', 'jardin', 'piscine', 'cuisine_equiped', 'terrasse', 'garage', 'quartier', 'price_de_m2', 'price']]
```

```python

# Fusionner les deux datasets
df_cleaned = pd.concat([df_mubawab_cleaned, df_avito_cleaned], ignore_index=True)

# Exporter le fichier final
df_cleaned.to_csv('data_cleaned.csv', index=False)
```
